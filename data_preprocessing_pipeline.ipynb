# Data Preprocessing Pipeline for Ride-Hailing Dataset



```python
# Import Libraries
import pandas as pd
import numpy as np
```

```python
# Load Dataset
df = pd.read_csv('renamed_actual_ridedata_sample.csv')
df.head()
```

```python
# Handle Missing Values
df.fillna(method='ffill', inplace=True)
df.fillna(method='bfill', inplace=True)
```

```python
# Check for duplicate rows
duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows: {len(duplicate_rows)}")

# Remove duplicates
df.drop_duplicates(inplace=True)

# Reset index after cleanup
df.reset_index(drop=True, inplace=True)
```

```python
# Filter only required columns for RL model
columns_to_keep = [
    'Pickup Location',
    'Request to Pickup',
    'Hour of Day',
    'Time of Day',
    'Month of Year',
]
df = df[[col for col in columns_to_keep if col in df.columns]]
df.to_csv('filtered_ridedata_sample.csv', index=False)
print('Filtered dataset saved as filtered_ridedata_sample.csv')
```

```python
# ðŸ’¾ Step 6: Save Preprocessed Dataset
df.to_csv('preprocessed_ride_hailing_dataset.csv', index=False)
print('Preprocessed data saved.')
```
